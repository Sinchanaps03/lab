```python
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
X, y = load_iris(return_X_y=True)
X = PCA(n_components=2).fit_transform(X)
plt.scatter(X[y==0,0], X[y==0,1], label='setosa')
plt.scatter(X[y==1,0], X[y==1,1], label='versicolor')
plt.scatter(X[y==2,0], X[y==2,1], label='virginica')
plt.legend()
plt.show()
```

```python
from sklearn.datasets import fetch_olivetti_faces
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# Load data
data = fetch_olivetti_faces()
X, y = data.data, data.target

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

# Train and predict
model = GaussianNB()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

# Accuracy
print("Accuracy:", accuracy_score(y_test, y_pred))

# Show few images with predictions
for i in range(5):
    plt.imshow(X_test[i].reshape(64, 64), cmap='gray')
    plt.title(f"True: {y_test[i]}, Pred: {y_pred[i]}")
    plt.axis('off')
    plt.show()
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Load and scale data
data = load_breast_cancer()
X = StandardScaler().fit_transform(data.data)

# KMeans clustering
kmeans = KMeans(n_clusters=2, random_state=0)
y_kmeans = kmeans.fit_predict(X)

# PCA for 2D visualization
X_pca = PCA(n_components=2).fit_transform(X)

# Plot
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_kmeans, cmap='Set1')
plt.title("K-Means Clustering")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```
