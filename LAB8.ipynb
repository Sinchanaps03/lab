```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
```

```python
# Load data
X, y = load_breast_cancer(return_X_y=True)

# Train model
model = DecisionTreeClassifier()
model.fit(X, y)

# Predict a new sample
sample = [X[0]]
print("Prediction:", ["Malignant", "Benign"][model.predict(sample)[0]])

# Plot decision tree
plt.figure(figsize=(25, 15))  # Makes the tree larger and more readable
plot_tree(model, filled=True, fontsize=8)  # Slightly bigger font
plt.show()
```
